# Step 2: Use Hugging Face's Free AI for Content Generation

```python
from transformers import pipeline
generator = pipeline("text-generation", model="bigscience/bloom-560m")

def generate_brochure_content(product_name, industry, key_features):
    prompt = f"Create a professional marketing brochure for {product_name} in the {industry} industry. Key features: {key_features}. Write an engaging introduction, key benefits, and a call to action."

    response = generator(prompt, max_length=200)[0]["generated_text"]
    return response

```

```python
from transformers import pipeline
from fpdf import FPDF

# Load a free text-generation model
generator = pipeline("text-generation", model="bigscience/bloom-560m")

# Function to generate marketing content
def generate_brochure_content(collection_name, industry, key_features):
    prompt = (f"Create a professional marketing brochure for {collection_name} in the {industry} industry. "
              f"This collection is designed for modern women who love elegance and comfort. "
              f"Key features include: {', '.join(key_features)}. "
              f"Write an engaging introduction, key benefits, and a call to action.")

    response = generator(prompt, max_length=250, truncation=True)[0]["generated_text"]
    return response

# Function to create a PDF brochure with an image
def create_pdf(content, image_path, filename="Summer_Collection_Brochure.pdf"):
    pdf = FPDF()
    pdf.set_auto_page_break(auto=True, margin=15)
    pdf.add_page()

    # Ensure the image is in a valid format (JPG, PNG, GIF)
    pdf.image(image_path, x=10, y=10, w=190)  # Adjust width & position
    pdf.ln(60)  # Space after image

    # Title
    pdf.set_font("Arial", "B", 16)
    pdf.cell(200, 10, "SUMMER VOLUME ONE", ln=True, align="C")
    pdf.ln(10)

    # Content
    pdf.set_font("Arial", size=12)
    pdf.multi_cell(0, 10, content.encode('latin-1', 'replace').decode('latin-1'))  # Fix Unicode error

    # Save PDF
    pdf.output(filename, "F")
    print(f"✅ Marketing brochure PDF created successfully: {filename}")

    return filename  # Return file path for download

# === Customize Your Brochure ===
collection_name = "Summer Volume One"
industry = "Women's Fashion"
key_features = ["Breathable Fabrics", "Elegant Maxi Dresses", "Trendy Tunics", "Lightweight Co-ord Sets", "Exclusive Limited Designs"]

# Generate content
content = generate_brochure_content(collection_name, industry, key_features)

# Provide a valid image file (JPEG, PNG, or GIF)
image_path = "/content/ITTEHAD_Lawn_Collection_-_Eastern_Fashion_600x.jpg"  # Replace with your image file path (NOT a PDF)

# Create the PDF
pdf_file = create_pdf(content, image_path)

# === Make it Downloadable ===
import shutil

shutil.move(pdf_file, "/content/Summer_Collection_Brochure.pdf")  # Move to a public location
print("📥 Download your brochure here: /content/Summer_Collection_Brochure.pdf")
```
